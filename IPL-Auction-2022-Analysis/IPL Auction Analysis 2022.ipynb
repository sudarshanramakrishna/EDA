```python
#replacing nan values in 2021 Sqaud column by not participated
ipl_data['2021 Squad']=ipl_data['2021 Squad'].fillna('Not Participated')
```

```python
ipl_data['status']=ipl_data['Team'].replace(teams,'Sold')
```

```python
#Number of player sold and unsold in auction
plt.figure(figsize=(5,5))
fig=sns.countplot(ipl_data['status'],palette=['Red','Blue'])
plt.xlabel('sold or unsold')
plt.ylabel('Number of players')
plt.title('Sold vs Unsold',fontsize=15)
plt.plot()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width()/2.,  p.get_height()), ha = 'center',
                 va = 'center', xytext=(0,4),textcoords='offset points')
```

```python
#printing sold and unsold players
ipl_data.groupby('status')['Player'].count()
```

```python
#total number of players bought by each team

plt.figure(figsize=(20,10))
fig=sns.countplot(ipl_data[ipl_data['Team']!='Unsold']['Team'])
plt.xlabel('Team Names')
plt.ylabel('Number of Players')
plt.title('Players bought by each Team', fontsize=15)
plt.xticks(rotation=70)
plt.plot()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.0f'),(p.get_x() + p.get_width()/2., p.get_height()), va = 'center',
                   ha = 'center', xytext=(0,4), textcoords='offset points')
```

```python
ipl_data['retention']=ipl_data['Base Price']
ipl_data.head()
```

```python
#lets see unique items in the retention columns so that we can replace 
ipl_data['retention'].unique()
```

```python
ipl_data['retention'].replace(['2 Cr', '40 Lakh', '20 Lakh', '1 Cr', '75 Lakh',
       '50 Lakh', '30 Lakh', '1.5 Cr'], 'From Auction', inplace=True)
ipl_data.head()
```

```python
ipl_data['Base Price'].replace(['Draft Pick','Retained'], 0, inplace=True)
ipl_data.head()
```

```python
#splittig base price into 2 parts : one as with value and another as its unit
ipl_data['base_price'] = ipl_data['Base Price'].apply(lambda x: str(x).split(' ')[0])
ipl_data['base_price_unit'] = ipl_data['Base Price'].apply(lambda x: str(x).split(' ')[-1])
ipl_data.head()  
```

```python
#total number of retained and bought
ipl_data.groupby(['Team','retention'])['retention'].count()[:-1]
```

```python
#players in each team based on different types of players  
#hue used for coloring based in type of player
plt.figure(figsize=(20,10))
fig=sns.countplot(ipl_data[ipl_data['Team']!='Unsold']['Team'], hue = ipl_data['TYPE'])
plt.title('Players in each team')
plt.xlabel('Team Names')
plt.ylabel('Number of Players')
plt.xticks(rotation=50)
```

```python
#Highest amount spent on a single player by each team
ipl_data[ipl_data['retention']=='From Auction'].groupby(['Team'])['COST IN ₹ (CR.)'].max()[:-1].sort_values(ascending=False)
```

```python
#player retained at maximum price
ipl_data[ipl_data['retention']=='Retained'].sort_values(by='COST IN ₹ (CR.)',ascending=False)
```

```python
#top 5 bowlers
ipl_data[(ipl_data['retention']=='From Auction') & (ipl_data['TYPE']=='BOWLER')].sort_values(by='COST IN ₹ (CR.)',ascending=False)
```

```python
#top 5 Batters
ipl_data[(ipl_data['retention']=='From Auction') & (ipl_data['TYPE']=='BATTER')].sort_values(by='COST IN ₹ (CR.)',ascending=False)
```

```python
#top 5 All rounders
ipl_data[(ipl_data['retention']=='From Auction') & (ipl_data['TYPE']=='ALL-ROUNDER')].sort_values(by='COST IN ₹ (CR.)',ascending=False)
```
